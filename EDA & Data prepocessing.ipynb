```python
#defined my input and target columns
X = df.drop("price_log", axis=1)
y = df["price_log"]
```

```python
#split the data into training sets and validation sets
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42
)
```

```python
#validate the split
print("Training set size:", X_train.shape)
print("Validation set size:", X_val.shape)
```

```python
#combine and save to parquet
train_df = X_train.copy()
train_df["price"] = y_train.values

val_df = X_val.copy()
val_df["price"] = y_val.values

train_df.to_parquet("training_data.parquet", index=False)
val_df.to_parquet("val_data.parquet", index=False)

```

```python
#outlier detection on prices
Q1 = df["price_log"].quantile(0.25)
Q3 = df["price_log"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df["price_log"] < Q1 - 1.5 * IQR) |
    (df["price_log"] > Q3 + 1.5 * IQR)
]

len(outliers)
```

```python
#outlier detection for area column
Q1 = df["area"].quantile(0.25)
Q3 = df["area"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df["area"] < Q1 - 1.5 * IQR) |
    (df["area"] > Q3 + 1.5 * IQR)
]

len(outliers)
```

```python
#binary encoding
binary_cols = [
    "mainroad",
    "guestroom",
    "basement",
    "hotwaterheating",
    "airconditioning",
    "prefarea"
]

binary_map = {"no": 0, "yes": 1}
# Training set
for col in binary_cols:
    train_df[col] = train_df[col].map(binary_map)

# Validation set
for col in binary_cols:
    val_df[col] = val_df[col].map(binary_map)

```

```python
numerical_cols = ["area", "bedrooms", "bathrooms", "stories", "parking"]
X_train = train_encoded.drop(["price"], axis=1)
y_train = train_encoded['price']


X_val = val_encoded.drop(["price"], axis=1)
y_val = val_encoded['price']
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize scaler
scaler = StandardScaler()

# Fit scaler on training set and transform training set
X_train_scaled = X_train.copy()  # keep original intact
X_train_scaled[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])

# Transform validation set using the same scaler
X_val_scaled = X_val.copy()
X_val_scaled[numerical_cols] = scaler.transform(X_val[numerical_cols])
```

```python
#done building the prepocessing now i can build an actual model and test after optimize the model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize model
lr_model = LinearRegression()

# Train model
lr_model.fit(X_train_scaled, y_train)
```

```python
#make a predicction on train set
y_train_pred = lr_model.predict(X_train_scaled)
#y_train_pred_price = np.exp(y_train_pred)
y_train_pred
```

```python
y_train
```

```python
# Predict on validation data
y_val_pred_log = lr_model.predict(X_val_scaled)
y_val_pred_price = np.exp(y_val_pred_log)
y_val_pred_log
```

```python
y_val
```

```python
#convert back to actual prices
y_train_price = np.exp(y_train)
y_train_pred_price = np.exp(y_train_pred)
#train performance
rmse = np.sqrt(mean_squared_error(y_train_price, y_train_pred_price))
mae = mean_absolute_error(y_train_price, y_train_pred_price)
r2 = r2_score(y_train_price, y_train_pred_price)

print(f"RMSE: {rmse:.4f}")
print(f"MAE : {mae:.4f}")
print(f"R²  : {r2:.4f}\n")

#evaluate_log(y_train, y_train_pred_log, "Training Set")



```
